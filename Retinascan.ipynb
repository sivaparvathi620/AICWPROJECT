```python
import os
import tensorflow as tf

# Settings
IMG_SIZE = (224, 224)
BATCH_SIZE = 32

# నీ స్క్రీన్‌షాట్ ప్రకారం కచ్చితమైన పాత్ ఇది:
retina_train_path = os.path.join('dataset', 'retina scan', 'content', 'Diabetic_Balanced_Data', 'train')

if not os.path.exists(retina_train_path):
    print(f"❌ Error: {retina_train_path} దొరకలేదు!")
    # ఒకవేళ దొరకకపోతే, మెయిన్ ఫోల్డర్ లో ఏముందో చూద్దాం
    print("చెక్ చేయి బాబు, ఈ ఫోల్డర్ ఉందా?:", os.listdir(os.path.join('dataset', 'retina scan', 'content', 'Diabetic_Balanced_Data')))
else:
    # ఇక్కడ 2 classes అని రావాలి
    train_retina = tf.keras.utils.image_dataset_from_directory(
        retina_train_path,
        validation_split=0.2,
        subset="training",
        seed=123,
        image_size=IMG_SIZE,
        batch_size=BATCH_SIZE
    )

    val_retina = tf.keras.utils.image_dataset_from_directory(
        retina_train_path,
        validation_split=0.2,
        subset="validation",
        seed=123,
        image_size=IMG_SIZE,
        batch_size=BATCH_SIZE
    )
    
    print("✅ Retina Dataset Loaded with 2 Classes!")
```

```python
# 1. Retina Model Architecture
model_retina = models.Sequential([
    layers.Rescaling(1./255, input_shape=(224, 224, 3)),
    layers.Conv2D(32, (3, 3), activation='relu'),
    layers.MaxPooling2D(),
    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.MaxPooling2D(),
    layers.Flatten(),
    layers.Dense(128, activation='relu'),

    layers.Dense(5, activation='softmax') 
])

model_retina.compile(
    optimizer='adam', 
    loss='sparse_categorical_crossentropy', 
    metrics=['accuracy']
)
print("✅ Retina Model Compiled Successfully!")
# 2. Train the Retina Model

model_retina.fit(train_retina, validation_data=val_retina, epochs=5)


model_retina.save('retina_model.h5')
print("✅ Retina Model Saved Successfully!")
```
